```python
!pip install seaborn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('medical_examination.csv')
df
```

```python
bmi = (df['weight']/ (df['height']/ 100)**2)
bmi
```

```python
a = []
for x in bmi:
    if x > 25 :
        a.append(1) 
    else:
        a.append(0) 
df['overweight'] = pd.Series(a)
df
```

```python
df['cholesterol'][df['cholesterol'] == 1] = 0
df['gluc'][df['gluc'] == 1] = 0
df
```

```python
df['cholesterol'][df['cholesterol'] > 1] = 1
df['gluc'][df['gluc'] > 1] = 1
df
```

```python
df_new = pd.melt(df,id_vars=['cardio'], value_vars=['cholesterol','gluc','smoke','alco','active','overweight'])
df_new
```

```python
sns.catplot(data=df_new, x='variable', hue='value', kind='count', col='cardio')
```

```python
df
```

```python
df_new = df[df['ap_lo'] <= df['ap_hi']]
df_new = df_new[df['height'] >= df['height'].quantile(0.025)]
df_new = df_new[df['height'] <= df['height'].quantile(0.975)]
df_new = df_new[df['weight'] >= df['weight'].quantile(0.025)]
df_new = df_new[df['weight'] <= df['weight'].quantile(0.975)]
df_new
```

```python
corr = df_new.corr()
corr
```

```python
f, ax = plt.subplots(figsize=(11, 9))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True)
```
